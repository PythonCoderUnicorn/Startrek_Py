# Startrek Python -  Modules

Modules are a bunch of python files that provide functionality, (like NumPy mentioned before for Arrays) that can be included into your application.

Modules are like books in a library, each book is a module which contains information for how to use a certain function or capability. Python libraries are code libraries.

- SURPISE, you have already made your own by creating functions! You first made a function to say hello/hi and if you rather not type it again for another program you can just import the file name {excluding the .py part} (tax free!)

```python
# module for clearing holosuite program history 
# and ending program

def end_program(program):
    holosuite_hx = ["Risa",'Orion','Cardassia']
    holosuite_hx.clear()
    print(": {} program ended :".format(program))
    print('holosuite history: {}'.format(holosuite_hx))
    
end_program('Risa')

# if using a text editor save file as : holosuite.py
# if using Notebook, create a new Python file, add 
# function above and save it as: holosuite.py 
```

## Module Alias

Using modules is awesome but typing the full name of the module everytime you need to call on it (reference it) to get a function 
is slow, tiring and not efficient. 

As mentioned before, module NumPy exists but whenever you want to use it don't type numpy more than once. Module alias is standard practice in Python, how to make an alias for module is easy! Just use the word 'as' followed by abreviation of module.

- <code>import numpy as np</code>  Now you can say <code>np.array()</code> or whatever you need/ want to use from NumPy module

- most alias naming follows a pattern of first letter of each word of module, but the alias abreviation is whatever you want and don't forget the name as you will get errors if name/ abreviations don't match in your code (seaborn as sns but later you have sb => error)

Now let's create a alias for the holosuite.py module

<code>import holosuite as hs</code>

Create an object to call module's dict key 'name'

<code>game = hs.holo_games['name']</code>

<code>print(game)</code>
